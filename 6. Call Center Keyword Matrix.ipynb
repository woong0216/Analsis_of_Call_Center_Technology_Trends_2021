```python
import numpy as np
import pandas as pd
import string
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk import bigrams
from nltk.corpus import wordnet as wn
from nltk.stem import WordNetLemmatizer
import itertools
import re
```

```python
Corpus = pd.read_csv('./data/call center period 3.csv')
```

### 소문자화

```python
step_a=[]
for sentence in Corpus['abstract']:
    sentence=sentence.lower()
    step_a.append(sentence)
```

### 정규화

```python
step_b=[]
for sentence in step_a:
    document = re.sub(r'[^ ㄱ-ㅣ가-힣A-Za-z]', '', sentence)
    step_b.append(document)
```

### stopwords 전처리

```python
import spacy
sp = spacy.load('en_core_web_sm')
spacy_stopwords=sp.Defaults.stop_words
```

```python
stop_list = set(stopwords.words('english'))
```

```python
from gensim.parsing.preprocessing import STOPWORDS
stopwords = STOPWORDS.union(spacy_stopwords, stop_list)
```

```python
step_c=[]

for sentence in step_b:
    sentence = sentence.split()
    step_c_1=[]
    for cor in sentence:
        if cor in stopwords:
            continue
        step_c_1.append(cor)
    step_c_1=" ".join(step_c_1)    
    step_c.append(step_c_1)
```

### lemmatizer

```python
n = WordNetLemmatizer()
lemmatizer_word=[]
for sentence in step_c:
    sentence=word_tokenize(sentence)
    word=[]
    for a in sentence:        
        a=n.lemmatize(a)
        word.append(a)
    word=pos_tag(word)
    lemmatizer_word.append(word)
```

### 명사만 추출

```python
sentence_a=[]

for sentence in lemmatizer_word:
    noum_set = []
    for t in sentence:
        word_li=[]
        #if (t[1] == 'VB' or t[1] == 'VBD' or t[1] == 'VBG' or t[1] == 'VBN' or t[1] == 'VBP' or t[1] == 'VBZ') or (len(t[0])<=1):
        if (t[1]=='NNP'):
            continue
        else:
            word_li.append(t[0])
        noum_set.append(word_li)
    sentence_a.append(noum_set)
```

### 2글자 이상만 추출

```python
len(sentence_a[0][0][0])
```

```python
sentence_b=[]

for sentence in sentence_a:
    row=[]
    for lists in sentence:
        rows=[]
        for a in lists:
            if len(a)<2:
                continue
            else:
                rows.append(a)
            row.append(rows)
    sentence_b.append(row)
```

### 기타 추출

```python
sentence_c=[]

for sentence in sentence_b:
    row=[]
    for lists in sentence:
        rows=[]
        for a in lists:
            a=a.replace(" is "," ")
            a=a.replace(" the "," ")
            a=a.replace(" with "," ")
            a=a.replace(" at "," ")
            a=a.replace(" to "," ")
            a=a.replace(" of "," ")
            a=a.replace(" by "," ")
            a=a.replace(" also "," ")
            a=a.replace(" this "," ")
            a=a.replace(" there "," ")
            a=a.replace('introduce',"")
            a=a.replace('result','')
            a=a.replace('method','')
            a=a.replace('related','')
            a=a.replace('work','')
            a=a.replace('conclusion','')
            rows.append(a)
            row.append(rows)
    sentence_c.append(row)
```

```python
Corpus['new_abstract']=sentence_c
Corpus.rename(columns = {'Unnamed: 0' : 'document'}, inplace = True)
```

### 재정의

```python
Corpus.head()
```

```python
cpc_set1=[]
for data in Corpus['new_abstract']:
    cpcs=[]
    for cpc in data:
        cpc=str(cpc)
        cpc=cpc.replace("'","")
        cpc=cpc.replace("[","")
        cpc=cpc.replace("]","")
        cpc=cpc.strip()
        cpcs.append(cpc)
    cpc_set1.append(cpcs)
```

### co-occurrence

```python
def generate_co_occurrence_matrix(corpus):
    vocab = set(corpus)
    vocab = list(vocab)
    vocab_index = {word: i for i, word in enumerate(vocab)}
 
    bi_grams = list(bigrams(corpus))
 
    bigram_freq = nltk.FreqDist(bi_grams).most_common(len(bi_grams))
 
    co_occurrence_matrix = np.zeros((len(vocab), len(vocab)))
 
    for bigram in bigram_freq:
        current = bigram[0][1]
        previous = bigram[0][0]
        count = bigram[1]
        pos_current = vocab_index[current]
        pos_previous = vocab_index[previous]
        co_occurrence_matrix[pos_current][pos_previous] = count
    co_occurrence_matrix = np.matrix(co_occurrence_matrix)
 
    return co_occurrence_matrix, vocab_index
 
text_data = cpc_set1 # 바꾸기

data = list(itertools.chain.from_iterable(text_data))
matrix, vocab_index = generate_co_occurrence_matrix(data)
 
data_matrix = pd.DataFrame(matrix, index=vocab_index, columns=vocab_index)

print(data_matrix)
```

```python
keyword_co_occurrence = pd.DataFrame(data_matrix)
```
