# patent의 group id 추출

- patent id와 group id가 1대1로 되어있어서 1대N으로 변환해야 함

```python
dataset2 = dataset2[['patent_id','subgroup_id']]
```

```python
dataset2.head()
```

- cpc / 상위까지 추출

```python
cpc_set=[]
for subgroup in dataset2['subgroup_id']:
    cpc=subgroup.split('/')[0]
    cpc_set.append(cpc)
```

```python
cpc_set
```

```python
dataset2['subgroup_id']=cpc_set
```

```python
dataset2.head()
```

- patent id 그룹별로 group id를 ','를 통해 join화

```python
dataset2=dataset2.groupby('patent_id').agg({'subgroup_id':', '.join}).reset_index()
```

```python
dataset2
```

# dataset1 + dataset2

```python
dataset = dataset[['id','date','abstract','title']]
```

```python
dataset.rename(columns={'id':'patent_id'}, inplace=True)
```

```python
dataset.head()
```

- patent_id를 통해 2가지 데이터 셋 병합

```python
dataset3 = pd.merge(dataset, dataset2, on='patent_id',how='inner')
```

```python
dataset3.head()
```

- 중복된 cpc 제거

```python
cpc_list=[]
for cpc in dataset3['subgroup_id']:
    split_cpc=cpc.split(', ')
    split_cpc=list(set(split_cpc))
    cpc_list.append(split_cpc)
```

```python
dataset3['cpc_list']=cpc_list
```

```python
dataset3.head()
```

```python
del dataset3['subgroup_id']
```

```python
dataset3.head()
```

# 데이터 저장

```python
dataset3.to_csv('./data/dataset.csv')
```
