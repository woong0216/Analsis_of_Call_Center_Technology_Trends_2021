```python
import pandas as pd
df = pd.read_csv('./data/call center period 3.csv')
```

```python
df.head()
```

```python
cpc_set = []

for cpc in df['cpc_list']:
    cpc=cpc.replace("'","")
    cpc=cpc.replace("[","")
    cpc=cpc.replace("]","")
    split_cpc = cpc.split(',')
    cpc = ' '.join(split_cpc).split()
    each_set = []
    for each in cpc:
        each_set.append(each)
    cpc_set.append(each_set)
```

```python
df['cpc_set']=cpc_set
```

```python
def flat_list(array): 
    a=[]
    for i in array:
        if type(i) == type(list()):
            a+=(flat_list(i))       
        else:
            a.append(i)
    return a

print(flat_list(df.cpc_set))
flat_list_1 = flat_list(df.cpc_set)
```

```python
list_a = []
flat_list_1[0]
list_a.append(flat_list_1[0].split(','))
result = []
for item in list_a:
    result.extend(item)
result= ' '.join(result).split()
result
```

```python
list_a = []
for x in range(0,len(flat_list_1)):
    list_a.append(flat_list_1[x].split(','))
    result = []
    for item in list_a:
        result.extend(item)
    result= ' '.join(result).split()
    
result
```

```python
from collections import Counter
cpc_code_list = flat_list(result)
cpc_code = Counter(cpc_code_list)
cpc_code
```

```python
from collections import OrderedDict
def sort_by_key(t):
    return t[1]
for k, v in OrderedDict(sorted(cpc_code.items(), key=sort_by_key, reverse=True)).items():
    print(k, v)
```
