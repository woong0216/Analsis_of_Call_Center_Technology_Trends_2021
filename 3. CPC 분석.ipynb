### 상위+하위 CPC

```python
cpc_set2=[]
for data in dataset['cpc_list']:
    data=data.split(',')
    cpcs=[]
    for cpc in data:
        cpc=cpc.replace("'","")
        cpc=cpc.replace("[","")
        cpc=cpc.replace("]","")
        cpc=cpc.strip()
        cpcs.append(cpc)
    cpc_set2.append(cpcs)
```

# CPC co-occurrence

## 상위 cpc

```python
import numpy as np
import nltk
from nltk import bigrams
import itertools
import pandas as pd
 
def generate_co_occurrence_matrix(corpus):
    vocab = set(corpus)
    vocab = list(vocab)
    vocab_index = {word: i for i, word in enumerate(vocab)}
 
    bi_grams = list(bigrams(corpus))
 
    bigram_freq = nltk.FreqDist(bi_grams).most_common(len(bi_grams))
 
    co_occurrence_matrix = np.zeros((len(vocab), len(vocab)))
 
    for bigram in bigram_freq:
        current = bigram[0][1]
        previous = bigram[0][0]
        count = bigram[1]
        pos_current = vocab_index[current]
        pos_previous = vocab_index[previous]
        co_occurrence_matrix[pos_current][pos_previous] = count
    co_occurrence_matrix = np.matrix(co_occurrence_matrix)
 
    return co_occurrence_matrix, vocab_index
 
text_data = cpc_set1 # 바꾸기

data = list(itertools.chain.from_iterable(text_data))
matrix, vocab_index = generate_co_occurrence_matrix(data)
 
data_matrix = pd.DataFrame(matrix, index=vocab_index, columns=vocab_index)

print(data_matrix)
```

```python
cpc_co_occurrence = pd.DataFrame(data_matrix)
```

```python
cpc_co_occurrence
```

```python
cpc_co_occurrence.to_csv("./data/co-occurrence call center period 3 상위.csv",encoding='utf8')
```
