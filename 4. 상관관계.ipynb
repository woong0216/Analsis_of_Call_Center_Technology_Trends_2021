## 딕셔너리 화

### node 딕셔너리화

```python
ID = node['Id'].values
modularity = node['modularity_class'].values
```

```python
dict = {'Source':ID, 'source modularity':modularity}
node_dict1 = pd.DataFrame(dict)
```

```python
dict = {'Target':ID, 'target modularity':modularity}
node_dict2 = pd.DataFrame(dict)
```

## edge 딕셔너리화

#### edge source 딕셔너리화

```python
Source = edge['Source'].values
Target = edge['Target'].values
```

```python
dict2 = {'Source':Source, 'Target':Target}
edge_dict = pd.DataFrame(dict2)
```

## 병합

### source 병합

```python
merge = pd.merge(edge_dict, node_dict1, on ='Source', how='right')
```

```python
merge
```

```python
del merge['Source']
```

```python
merge2 = pd.merge(merge, node_dict2, on ='Target', how='right')
```

```python
merge2
```

```python
del merge2['Target']
```

```python
merge2
```

```python
merge2['source modularity'].unique()
```

```python
df11=merge2.groupby('source modularity')['target modularity'].apply(lambda x: (x=='one').sum()).reset_index(name='count')
print (df11)
```

```python
merge2['source modularity'] = merge2['source modularity'].astype('category')
df1 = merge2[merge2['target modularity'] == 'one'].groupby(['source modularity']).size().reset_index(name='count') 
```

```python
df2 = merge2.groupby(['source modularity', 'target modularity']).size().reset_index(name='count') 
print(df2)
```

```python
df3 = merge2.groupby(['source modularity', 'target modularity']).size().unstack(fill_value=0)
print (df3)
```
